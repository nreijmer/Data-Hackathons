```python
import pandas as pd
import math as math
import numpy as np

train = pd.read_csv('C:/Hackathon/Black Friday/data/train.csv')
test = pd.read_csv('C:/Hackathon/Black Friday/data/test.csv')

print 'data imported'
```

```python
train['Stay_In_Current_City_Years']=train['Stay_In_Current_City_Years'].replace('4+', '4')
train = train.replace({'Gender': {'F':'1', 'M':'0'},
                              'City_Category': {'A':'0', 'B':'1', 'C':'2'},
                               'Age': {'0-17':'10','18-25':'20','26-35':'30','36-45':'40','46-50':'48','51-55':'53','55+':'60'}                         
                               
                              })
```

```python
test['Stay_In_Current_City_Years']=test['Stay_In_Current_City_Years'].replace('4+', '4')
test = test.replace({'Gender': {'F':'1', 'M':'0'},
                              'City_Category': {'A':'0', 'B':'1', 'C':'2'},
                               'Age': {'0-17':'10','18-25':'20','26-35':'30','36-45':'40','46-50':'48','51-55':'53','55+':'60'}                         
                               
                              })
```

```python
inputVars = train[['Gender', 'City_Category','Age', 'Occupation', 'Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category_1']]
outputVars = test[['Gender', 'City_Category','Age', 'Occupation', 'Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category_1']]
```

```python
print 'importeren decision tree libraries'

from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(inputVars, train['Purchase'])
```

```python
print "voorspellingen maken"

prediction = clf.predict(outputVars)

prediction
```
